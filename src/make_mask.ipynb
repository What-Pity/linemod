```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import json
from pathlib import Path

folder=Path("../data/multiple_items")
```

```python
template=cv2.imread(folder/'all.jpg')
plt.imshow(template[:,:,::-1])
```

```python
with open(folder/'all.json') as f:
    boundary = np.array(json.load(f)["shapes"][0]["points"],dtype=np.int32)

mask=np.zeros((template.shape[0],template.shape[1]),dtype=np.uint8)
cv2.fillPoly(mask,[boundary],255)
plt.imshow(mask,cmap="gray")
```

```python
cv2.imwrite(str(folder/"arrow.jpg"),mask)
```
